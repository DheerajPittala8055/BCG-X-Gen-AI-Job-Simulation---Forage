# <center > Financial Analysis of Microsoft, Tesla, and Apple (Last 3 Fiscal Years)
## Objective
This notebook extracts key financial data from 10-K filings for Microsoft, Tesla, and Apple, then analyzes trends to provide insights for an AI-powered financial chatbot.

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Importing the 10K-data for the past 3 years since 2024.

df = pd.read_csv(r"C:\Users\dp1257\Downloads\10-K Data.csv")

df.head()
```

```python
# Compute year-over-year percentage change

df['Revenue Growth (%)'] = df.groupby('Company')['Total Revenue'].pct_change() * 100
df['Net Income Growth (%)'] = df.groupby('Company')['Net Income'].pct_change() * 100
df['Assets Growth (%)'] = df.groupby('Company')['Total Assets'].pct_change() * 100
df['Liabilities Growth (%)'] = df.groupby('Company')['Total Liabilities'].pct_change() * 100
df['Cash Flow Growth (%)'] = df.groupby('Company')['Cash Flow from Operating Activities'].pct_change() * 100

df
```

```python
df.fillna(0, inplace=True)

df
```

```python
df.to_csv('final_data_report.csv')
```

```python
summary = df.groupby('Company').agg({
    'Revenue Growth (%)': 'mean',
    'Net Income Growth (%)': 'mean',
    'Assets Growth (%)' : 'mean',
    'Liabilities Growth (%)' : 'mean',
    'Cash Flow Growth (%)' :'mean'    
}).reset_index()
```

```python
print('Year-By-Year Average Growth Rates(%) :-')
print('             OR')
print("Overall Growth/Fall rate for Apple, Microsoft and Tesla from Fiscal Year 2021 - 2023")

summary
```

```python
summary.to_csv('summary_report')
```

# Summary:

Apple has shown steady financial performance with a **0.30% increase in revenue** and a **2.12% rise in net income**, indicating strong profitability. While its **assets declined by 1.12%**, this was accompanied by a **0.57% reduction in liabilities**, suggesting controlled financial management. Additionally, Apple’s **1.33% increase in cash flow** highlights its operational efficiency, maintaining a healthy financial position despite minor asset reductions.  

Microsoft experienced an overall financial decline, with **revenue dropping by 6.66%** and **net income decreasing by 5.79%**, reflecting potential market challenges. The **10.33% reduction in assets** and **6.40% decline in liabilities** indicate a strategic financial shift, possibly involving asset restructuring or debt repayments. However, the **8.15% decline in cash flow** raises concerns about liquidity and future investment capabilities, making this a challenging fiscal period for the company.  

Tesla’s financial performance presents a mixed picture, with **revenue declining by 5.59%** and **assets dropping by 11.81%**, suggesting potential production slowdowns or restructuring. However, **net income surged by 31.13%**, indicating significant profitability improvements, likely driven by cost optimizations or better margins. Despite an **8.80% reduction in liabilities**, Tesla’s **cash flow remained relatively stable**, showing resilience in managing operational expenses while improving profitability.  

Comparing all three companies, **Apple stands out as the most stable**, with balanced financial growth and controlled liabilities. **Microsoft shows the most significant decline**, struggling with revenue, net income, and cash flow drops, indicating financial headwinds. **Tesla, despite declining revenue, demonstrates strong profitability**, suggesting strategic cost management. Overall, Apple and Tesla are in a stronger financial position, while Microsoft faces challenges requiring strategic adjustments.

## Visualizing Financial Trends
We will use line plots to visualize the year-over-year growth trends for each financial metric.

```python
# Plot revenue growth trends
plt.figure(figsize=(10, 5))
for company in df['Company'].unique():
    company_data = df[df['Company'] == company]
    plt.plot(company_data['Year'], company_data['Revenue Growth (%)'], marker='o', label=company)
plt.xlabel('Fiscal Year')
plt.ylabel('Revenue Growth (%)')
plt.title('Year-over-Year Revenue Growth')
plt.legend()
plt.grid()
plt.show()
```

```python
# Plot net income growth trends
plt.figure(figsize=(10, 5))
for company in df['Company'].unique():
    company_data = df[df['Company'] == company]
    plt.plot(company_data['Year'], company_data['Net Income Growth (%)'], marker='s', label=company)
plt.xlabel('Fiscal Year')
plt.ylabel('Net Income Growth (%)')
plt.title('Year-over-Year Net Income Growth')
plt.legend()
plt.grid()
plt.show()
```
